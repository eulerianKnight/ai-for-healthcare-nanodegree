````python
# Read the DICOM's pixel_array attribute
dcm_19 = pydicom.dcmread('dicom_00013659_019.dcm')
dcm_33 = pydicom.dcmread('dicom_00023075_033.dcm')
dcm_05 = pydicom.dcmread('dicom_00029579_005.dcm')
````

````python
# Visualize the image
for dcm in [dcm_19, dcm_33, dcm_05]:
    plt.figure()
    plt.imshow(dcm.pixel_array, cmap='gray')
````

````python
# Plot Histogram of Image Pixel Intensity Values
for dcm in [dcm_19, dcm_33, dcm_05]:
    plt.figure(figsize=(5, 5))
    plt.hist(dcm.pixel_array.ravel(), bins=256)
````

````python
# Find mean, standard deviation of images and standardize it
standardized_images = []
for dcm in [dcm_19, dcm_33, dcm_05]:
    print(f"Mean Intensity: {np.mean(dcm.pixel_array)}")
    print(f"Standard Deviation: {np.std(dcm.pixel_array)}")
    standardized_images.append((dcm.pixel_array.copy()))
````
